```python
import cv2
from pyzbar.pyzbar import decode

def read_qr_code(image_path):
    # 이미지 파일 읽기
    image = cv2.imread(image_path)

    # QR코드 디코딩
    decoded_objects = decode(image)

    # 디코딩된 정보 출력
    for obj in decoded_objects:
        print(f"타입: {obj.type}")
        print(f"데이터: {obj.data.decode('utf-8')}")
        print()

        # 디코딩된 QR코드를 이미지에 표시
        points = obj.polygon
        if len(points) == 4:
            pts = []
            for j in range(4):
                pts.append((points[j].x, points[j].y))

            pts = [(int(x), int(y)) for x, y in pts]
            pts = [(x, y) for x, y in pts]
            cv2.polylines(image, [pts], isClosed=True, color=(0, 255, 0), thickness=2)

    # 이미지 창에 디코딩된 QR코드 표시
    cv2.imshow("QR Code Reader", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 이미지 파일 경로
image_path = "path/to/your/image.jpg"  # 실제 파일 경로로 수정

# QR코드 판독 함수 호출
read_qr_code(image_path)
```

```python
import cv2
from pyzbar.pyzbar import decode

def read_qr_code(image_path):
    # 이미지 파일 읽기
    image = cv2.imread(image_path)

    # QR코드 디코딩
    decoded_objects = decode(image)

    # 디코딩된 정보 출력
    for obj in decoded_objects:
        qr_data = obj.data.decode('utf-8')
        print(f"타입: {obj.type}")
        print(f"데이터: {qr_data}")
        print()

        # 예상한 QR코드 데이터
        expected_qr_data = "예상한 QR 코드 데이터"  # 실제 데이터로 수정

        # 판별 결과 출력
        if qr_data == expected_qr_data:
            print("QR 코드 판별 결과: 일치합니다.")
        else:
            print("QR 코드 판별 결과: 일치하지 않습니다.")

        # 디코딩된 QR코드를 이미지에 표시
        points = obj.polygon
        if len(points) == 4:
            pts = [(int(x), int(y)) for x, y in points]
            pts = [(x, y) for x, y in pts]
            cv2.polylines(image, [pts], isClosed=True, color=(0, 255, 0), thickness=2)

    # 이미지 창에 디코딩된 QR코드 표시
    cv2.imshow("QR Code Reader", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 이미지 파일 경로
image_path = "path/to/your/image.jpg"  # 실제 파일 경로로 수정

# QR코드 판별 함수 호출
read_qr_code(image_path)
```
